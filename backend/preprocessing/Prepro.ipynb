```python
# week
df['week'] = df['week'].astype(int)
df['week_num'] = df['week'] % 100
df['year'] = df['week'] // 100
df.drop(["week"], axis=1, inplace=True)
```

```python
# month
# Function to calculate the month based on the week
def week_to_month(week):
    if 1 <= week <= 4:
        return "1"
    elif 5 <= week <= 9:
        return '2'
    elif 10 <= week <= 13:
        return '3'
    elif 14 <= week <= 17:
        return '4'
    elif 18 <= week <= 22:
        return '5'
    elif 23 <= week <= 26:
        return '6'
    elif 27 <= week <= 31:
        return '7'
    elif 32 <= week <= 35:
        return '8'
    elif 36 <= week <= 40:
        return '9'
    elif 41 <= week <= 44:
        return '10'
    elif 45 <= week <= 48:
        return '11'
    elif 49 <= week <= 52:
        return '12'
    else:
        return 'Invalid'  # Handle invalid week values

# Apply the function to create the 'month' column
df['month'] = df['week_num'].apply(week_to_month)
df['month'] = df['month'].astype(int)
```

```python
# performing one-hot encoding (month)
# df = pd.get_dummies(df, columns=['month'], prefix='month')
```

```python
# Location
# OnD to itinerary
# Create a dictionary to map unique combinations of "OnD" to category numbers
unique_combinations = df['OnD'].unique()
category_mapping = {combo: i + 1 for i, combo in enumerate(unique_combinations)}

# Map the "OnD" values to category numbers and create a new column "itinerary"
df['itinerary'] = df['OnD'].map(category_mapping)
```

```python
# df.drop(["Orig", "Dest", "OnD"], axis=1, inplace=True)
df.drop(["Dest", "OnD"], axis=1, inplace=True)
```

```python
# preprocess time further

# Departure time
# Extract dep_date
df["dep_date"] = pd.to_datetime(df.deptime_U, format="%d/%m/%Y").dt.day
# Extract dep_hour
df['dep_hour'] = [time_obj.hour for time_obj in df['deptime']]
df['dep_min'] = [time_obj.minute for time_obj in df['deptime']]

# drop
df.drop(["deptime_U", "deptime"], axis=1, inplace=True)
```

```python
# Arrive time
# Extract arr_date
df["arr_date"] = pd.to_datetime(df.arrtime_U, format="%d/%m/%Y").dt.day
# Extract arr_hour
df['arr_hour'] = [time_obj.hour for time_obj in df['arrtime']]
df['arr_min'] = [time_obj.minute for time_obj in df['arrtime']]

# drop
df.drop(["arrtime_U", "arrtime"], axis=1, inplace=True)
```
